```python
#Save code to calculate for individual state 

def state_format():
    
    for index, state in enumerate(mw_by_state['State']):
    
        current_index = index
        next_index = current_index + 1   
       
        state = mw_by_state[current_index:next_index].melt
        state = state()
        state['year-1'] = state['value'].shift(1)
        state['year-2'] = state['value'].shift(2)
        state['state']=(state['value'][0])
        state.rename(columns = {'variable':'year', 'value':'year-0'}, inplace = True)
        state = state.drop([0])

        state['year-1'].loc[1] = state['year-1'].loc[1].replace(str(state), 'NaN')
        state['year-2'].loc[2] = state['year-2'].loc[2].replace(str(state), 'NaN')

        print (state)
```

```python
all_states = mw_by_state.melt(id_vars = ['State'], value_vars =['1999','2000','2001','2002','2003','2004','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017'], var_name = 'year' ).groupby(by = ['State','year']).sum()
```

```python
all_states.head()
```

```python
all_states.rename(columns = {'value':'mw_year_0'}, inplace = True)
```

```python
all_states.head()
```

```python
all_states['mw_year_1'] = all_states['mw_year_0'].shift(1)
all_states['mw_year_2'] = all_states['mw_year_0'].shift(2)
```

```python
all_states.head()
```

```python
grand_total = all_states.reset_index()[all_states.reset_index()['State'] != 'Grand Total']
```

```python
grouped_states = all_states.groupby(by = 'year').sum().reset_index()
```

```python
grouped_states = grouped_states.drop(grouped_states.index[0:2])
```

```python
grouped_states
```

```python
year_0 = grouped_states['mw_year_0']
year_1 = grouped_states['mw_year_1']
year_2 = grouped_states['mw_year_2']

x = grouped_states['year']

plt.plot(x, year_0)
plt.plot(x, year_1)
plt.plot(x, year_2)

plt.xticks(np.arange(17, step = 2));
plt.xlabel('Year');
plt.ylabel('Megawatts (MW)');
plt.title('Time Series of Installed U.S. Wind Power Capacity');
```
