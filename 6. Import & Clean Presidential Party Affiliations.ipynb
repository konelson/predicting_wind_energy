```python
state_parties = {}
```

```python
def find_party_affiliation(i):
    
    political_dict = {}
    
    political_dict['Democratic'] = soup2016.find_all('td')[i+2].text
    political_dict['Republican'] = soup2016.find_all('td')[i+5].text
    
    state_parties[soup2016.find_all('td')[i].text] = political_dict
    
    return state_parties
```

```python
find_party_affiliation(156)
```

```python
for i in range(145,730,11):
    
    find_party_affiliation(i)
```
