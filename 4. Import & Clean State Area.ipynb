```python
from selenium import webdriver
import os

chromedriver = "/Applications/chromedriver"
os.environ["webdriver.chrome.driver"] = chromedriver

driver = webdriver.Chrome(chromedriver)
```

```python
driver.get('https://www.census.gov/geo/reference/state-area.html')
```

```python
driver.page_source
```

```python
selenium_soup = BeautifulSoup(driver.page_source,"html5lib")
```

```python
selenium_soup.find_all('td')[51].text,selenium_soup.find_all('td')[52].text
```

```python
selenium_soup.find_all('td')[68].text,selenium_soup.find_all('td')[69].text
```

```python
area_dict = {}

def find_areas():
    
    for i in range(51,900,17):
    
        area_dict[selenium_soup.find_all('td')[i].text] = selenium_soup.find_all('td')[i+1].text
   
```

```python
find_areas()
```

```python
#Total square miles of land and water areas per state

area_dict
```

```python
#Save final DataFrame as pickle file

with open ('state_area.pickle', 'wb') as to_write:
    pickle.dump(area_dict, to_write)
```
